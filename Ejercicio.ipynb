```python
import pandas as pd
data_juegos = pd.read_csv("juegos-ml.csv")
data_juegos
```

```python
# dividir el set de datos
# datos de entrada
X = data_juegos.drop(columns=["juegos"])

# obtener el set de datos del output
# datos de salida
y = data_juegos["juegos"]
data_juegos
```

```python
from sklearn.tree import DecisionTreeClassifier as DTREE
modelo = DTREE()
# con .values accedemos a los valores
# y eliminamos nombre de cabeceras
modelo.fit(X.values, y)
predicciones = modelo.predict([[14,0], [29,0], [30,0]])
predicciones
```

```python
# dividir datos de prueba y datos de entrenamiento
# asi medimos las predicciones

from sklearn.model_selection import train_test_split as TTS
from sklearn.metrics import accuracy_score as AS

# test_size = porcentaje de datos a utilizar
# tts retorna una tupla
# desempaquetar tupla

X_entrenar, X_prueba, y_entrenar, y_prueba = TTS(X.values, y, test_size=0.2)
modelo.fit(X_entrenar, y_entrenar)
predicciones = modelo.predict(X_prueba)
puntaje = AS(y_prueba, predicciones)
puntaje, predicciones
```

```python
# persistencia de modelos
import joblib as jb

modelo.fit(X.values, y)
jb.dump(modelo, "recomendador-juegos.joblib")
```

```python
modelo = jb.load("recomendador-juegos.joblib")
predicciones = modelo.predict([[14,0]])
predicciones
```
